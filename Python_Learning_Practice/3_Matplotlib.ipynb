```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# %matplotlib inline
```

```python
x = ['c++' , 'perl' , 'java' , 'python' , 'css' , 'matlab'] 
y = [70,40,75,95,65,50]
```

```python
plt.title("Popularity Programming Languages")
```

```python
plt.plot(x,y)
```

```python
plt.grid(True)
```

```python
plt.title('Popularity Programming Languages')
plt.plot(x,y,color='r',marker="s",linestyle='--',linewidth=3,markersize=13)
plt.ylim(0,100)
plt.grid(True)
plt.show()
```

```python
x = np.arange(-2,2,0.01)
y = np.sin(6 * np.pi * x)
y2 = np.sin(2 * np.pi * x) + 1.5
plt.plot(x,y)
plt.plot(x,y2)
plt.title("Mathematics & Statistical Diagram")
plt.grid(True)
plt.show()
```

```python
x = [4,6,8,10,12,14]
y = [12,8,10,4,7,20]
plt.bar(x,y)
bar_lable = ['java','c++','ruby','c','javascript','python']
plt.xticks(x,bar_lable)
plt.show()
```

```python
plt.title('Diagramm sample')
plt.bar([2017,2018,2019,2020],[90,140,120,90] , label = 'param1', width = 0.5)
plt.bar([2017,2018,2019,2020],[80,50,40,60] , label = 'param2', width = 0.5)
plt.xlabel('Year')
plt.ylabel('points')
plt.legend()
plt.show()
```

```python
data = np.random.randn(1000)
np.random.seed(42)
plt.hist(data,bins=20,color='gold')
plt.grid(True)
plt.title('Histogram sample')
plt.xlabel('points')
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
area = np.pi * (15 * np.random.rand(100)) ** 2
plt.scatter(x,y,s=area,c=colors,alpha=0.6)
plt.title('Scatter chart')
plt.show()
```
